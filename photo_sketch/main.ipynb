# Photo Sketching
In this project I will generate sketches in black-and-white based on preprovided color images

## Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
import PIL
```

## Read Image

First, let's read the stored image file using the imread function of cv2.

```python
img = cv2.imread("img/lion.jpeg")
plt.imshow(img)
```

Hmm... seems like the colors are off. This is because CV2 reads an image in BGR instead of RGB. Let's convert the image to the suitable format.

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
```

Great! Now the image appearance is as expected. Let's try converting this to grayscale format.

```python
image_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray)
print(image_gray)
```

Wow! That looks cool but it doesn't look gray... This is because we didn't set the color map to the correct one.

```python
plt.imshow(image_gray, cmap="gray")
```

## Edge Detection

Now we want to generate a sketch based on the provided image. We can do this by exhibiting the edges of the image. First, let's apply a Gaussian blur on the image:

```python
blurred = cv2.GaussianBlur(image_gray, (5, 5), 0)
plt.imshow(blurred, cmap="gray")
```

We will try three different parameter settings to control which lines appear.

```python
tight = cv2.Canny(blurred, 240, 250)
plt.imshow(tight, cmap="gray")
```

```python
mid = cv2.Canny(blurred, 30, 150)
plt.imshow(mid, cmap="gray")
```

```python
wide = cv2.Canny(blurred, 10, 200)
plt.imshow(wide, cmap="gray")
```
